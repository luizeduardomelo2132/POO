```python
class Forma:
  def calcular_area(self):
    pass

class Quadrado(Forma):
  def __init__(self, lado):
    self.lado = lado
  def area(self):
    return self.lado * self.lado

area_do_quadrado = Quadrado(int(input("digite: ")))
print("area do quadrado: ", area_do_quadrado.area())
```
